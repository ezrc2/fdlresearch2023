```python
alpha_values = np.arange(0.0, 1.0, 0.05)
alpha_values
```

```python
accuracy = []
precision = []
recall = []
f1_score = []
for alpha in alpha_values:
    print('alpha = ' + str(alpha))
    model = Classifier(alpha=alpha)
    report = model.train(csv_file='new_dataset.csv', model_type='SVM')
    accuracy.append(report['accuracy'])
    precision.append(report['macro avg']['precision'])
    recall.append(report['macro avg']['recall'])
    f1_score.append(report['macro avg']['f1-score'])
    print('-'*50)
```

```python
lst = [accuracy, precision, recall, f1_score]
names = ['accuracy', 'precision', 'recall', 'f1-score']
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
i = 0
for row in ax:
    for col in row:
        col.plot(alpha_values, lst[i])
        col.set_title(f'{names[i]} graph')
        col.set(xlabel='alpha', ylabel=names[i])
        i += 1
```
